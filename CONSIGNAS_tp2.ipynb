### 1. **Bucles**

   - **Actividad 1.1**: Escribir un programa que muestre el print de todo lo que el usuario introduzca hasta que el usuario escriba “salir” que terminará.
   
   - **Actividad 1.2**: Escribir un programa que pida al usuario un número entero y muestre por pantalla un triángulo rectángulo como el de más abajo con la cantidad de escalones indicados previamente (los numeros dentro del triangulo deben ser enteros aleatorios del 0 al 9).
* 1
* 3 1

```python

# todo: Resolución
# ! 1
def leer_entrada_hasta_salir():
   
    while True:
        entrada_usuario = input("Escribe algo (escribe 'salir' para terminar): ")
        if entrada_usuario.lower() == "salir":
            print("Programa terminado.")
            break
        print(f"Has escrito: {entrada_usuario}")

# Ejecuta la función
leer_entrada_hasta_salir()
```

```python

# todo: resolución

# ! 2
import random

def mostrar_triangulo_aleatorio():
    
    cantidad_filas = int(input("Ingrese el número de filas para el triángulo: "))
    
    for fila in range(1, cantidad_filas + 1):
        linea = []
        for _ in range(fila):
            numero_aleatorio = random.randint(0, 9)
            linea.append(str(numero_aleatorio))
        print(" ".join(reversed(linea)))

# Ejecuta la función
mostrar_triangulo_aleatorio()


```

### 2. **Listas y Tuplas**

   - **Actividad 1**: Escribir un programa que pida al usuario una frase y muestre por pantalla el número de veces que contiene cada vocal y la cantidad de espacios en blanco.

   - **Actividad 2**: Escribir un programa que cree una matriz (con numeros aleatorios del 1 al 100) usando las dimensiones de X e Y que el usuario deberia ingresar por consola, luego crea una tupla con la matriz transpuesta y por ultimo las muestra ambas por consola .

   - **Actividad 3**: Escribir un programa que almacene en una lista los siguientes precios, 50, 75, 46, 22, 80, 65, 8, y muestre por pantalla el menor y el mayor de los precios

```python

# todo: Resolución

# ! 1
def contar_vocales_y_espacios():
    
    frase = input("Ingrese una frase: ").lower()
    vocales = 'aeiou'
    conteo_vocales = {vocal: 0 for vocal in vocales}
    espacios = 0

    for caracter in frase:
        if caracter in conteo_vocales:
            conteo_vocales[caracter] += 1
        elif caracter == ' ':
            espacios += 1

    print("Conteo de vocales:")
    for vocal, cantidad in conteo_vocales.items():
        print(f"{vocal}: {cantidad}")
    print(f"Espacios en blanco: {espacios}")

# Ejecuta la función
contar_vocales_y_espacios()
```

```python

# todo: Resolución

# ! 2
import random

def crear_y_transponer_matriz():
   
    filas = int(input("Ingrese el número de filas: "))
    columnas = int(input("Ingrese el número de columnas: "))

    # Crear matriz original
    matriz = [[random.randint(1, 100) for _ in range(columnas)] for _ in range(filas)]

    # Transponer la matriz
    matriz_transpuesta = list(zip(*matriz))

    # Mostrar matrices
    print("\nMatriz original:")
    for fila in matriz:
        print(fila)
    
    print("\nMatriz transpuesta (como tupla):")
    for fila in matriz_transpuesta:
        print(fila)

# Ejecuta la función
crear_y_transponer_matriz()
```

### 3. **Diccionarios**

   - **Actividad 1**: Escribir un programa que cree un diccionario simulando una cesta de la compra. El programa debe preguntar el artículo y su precio y añadir el par al diccionario, hasta que el usuario decida terminar. Después se debe mostrar por pantalla la lista de la compra y el coste total, con el siguiente formato.

   | arts  |precio |
   |-------|-------|
   |art 1  |1      |
   |art 2  |2      |
   |art 3  |3      |
   |total  |6      |

   - **Actividad 2**: Escribir un programa que cree un diccionario de traducción español-inglés. El usuario introducirá las palabras en español e inglés separadas por dos puntos, y cada par <palabra> : <traducción> separados por comas (es decir, solo se debe usar **UNA VEZ** la función input para la carga de palabras). El programa debe crear un diccionario con las palabras y sus traducciones. Después pedirá una frase en español y utilizará el diccionario para traducirla palabra a palabra. Si una palabra no está en el diccionario debe dejarla sin traducir.

   - **Actividad 3**: Escribir un programa que reciba una cadena de caracteres y devuelva un diccionario con cada palabra que contiene y su frecuencia. Escribir otra función que reciba el diccionario generado con la función anterior y devuelva una tupla con la palabra más repetida y su frecuencia.

```python

# todo: Resolución

# ! 1

def cesta_compra():
   
    cesta = {}  # Diccionario para almacenar artículos y precios

    while True:
        articulo = input("Ingrese el nombre del artículo (o 'fin' para terminar): ").strip().lower()
        
        if articulo == "fin":
            break  # Se termina la carga

        if articulo == "":
            print("⚠️ El nombre del artículo no puede estar vacío.")
            continue

        try:
            precio = float(input(f"Ingrese el precio de '{articulo}': "))
            cesta[articulo] = precio  # Se guarda en el diccionario
        except ValueError:
            print("⚠️ Precio no válido. Ingrese un número.")
            continue

    # Mostrar los resultados finales
    if cesta:
        lista_articulos = ", ".join(cesta.keys())
        lista_precios = ", ".join(str(precio) for precio in cesta.values())
        total = sum(cesta.values())

        print(f"\narts: {lista_articulos}, total.")
        print(f"precio: {lista_precios}, {total}")
    else:
        print("La cesta está vacía.")

# Llamar a la función
cesta_compra()
```

```python

# todo: Resolución

# ! 2
def traduccion_espanol_ingles():
    
    # Entrada única con todos los pares de traducción
    entrada = input("Ingrese pares de traducción (español:ingles) separados por comas: ")
    diccionario_traduccion = {}
    
    # Procesar cada par de traducción
    pares = [par.strip() for par in entrada.split(",")]
    for par in pares:
        if ":" in par:
            espanol, ingles = par.split(":", 1)
            diccionario_traduccion[espanol.strip()] = ingles.strip()
    
    # Solicitar la frase en español a traducir
    frase = input("Ingrese una frase en español: ")
    palabras = frase.split()
    traduccion = []
    
    # Traducir palabra por palabra
    for palabra in palabras:
        traduccion.append(diccionario_traduccion.get(palabra, palabra))
    
    print("Traducción:", " ".join(traduccion))

# Ejecutar la función
traduccion_espanol_ingles()

```

```python

# todo: Resolución

# ! 3
def frecuencia_palabras(cadena):
   
    palabras = cadena.split()
    frecuencia = {}
    for palabra in palabras:
        palabra = palabra.lower()
        frecuencia[palabra] = frecuencia.get(palabra, 0) + 1
    return frecuencia

def palabra_mas_repetida(frecuencias):
    
    if not frecuencias:
        return None, 0
    palabra_maxima = max(frecuencias, key=frecuencias.get)
    return palabra_maxima, frecuencias[palabra_maxima]

# Ejemplo de uso de las funciones:
texto_usuario = input("Ingrese un texto: ")
diccionario_frecuencia = frecuencia_palabras(texto_usuario)
print("Frecuencia de palabras:", diccionario_frecuencia)
palabra, frecuencia = palabra_mas_repetida(diccionario_frecuencia)
print(f"La palabra más repetida es '{palabra}' con {frecuencia} ocurrencias.")
```

### 4. **Funciones**

- **Actividad 1**: Escribir una función que calcule el total de una factura tras aplicarle el IVA. La función debe recibir la cantidad sin IVA y el porcentaje de IVA a aplicar, y devolver el total de la factura. Si se invoca la función sin pasarle el porcentaje de IVA, deberá aplicar un 21%.

- **Actividad 2**: Escribir una función que reciba un número entero positivo y devuelva su factorial.


```python

# todo: Resolución

# ! 1
def calcular_total_factura(cantidad_sin_iva, porcentaje_iva=21):
   
    total_con_iva = cantidad_sin_iva * (1 + porcentaje_iva / 100)
    return total_con_iva

# Ejemplo de uso:
monto = float(input("Ingrese el monto sin IVA: "))
porcentaje = input("Ingrese el porcentaje de IVA (presione Enter para usar 21%): ")

# Convertir porcentaje si fue ingresado
if porcentaje:
    total = calcular_total_factura(monto, float(porcentaje))
else:
    total = calcular_total_factura(monto)

print(f"Total de la factura con IVA: {total:.2f}")
```

```python

# todo: Resolución

# ! 2
def calcular_factorial(numero):
    
    if numero < 0:
        raise ValueError("El número debe ser entero positivo.")
    
    factorial = 1
    for i in range(1, numero + 1):
        factorial *= i
    return factorial

# Ejemplo de uso:
try:
    numero_usuario = int(input("Ingrese un número entero positivo: "))
    resultado = calcular_factorial(numero_usuario)
    print(f"El factorial de {numero_usuario} es: {resultado}")
except ValueError as e:
    print("Error:", e)
```
